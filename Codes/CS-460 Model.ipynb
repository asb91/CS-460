```python
# IMPORTING REQUIRED MODULES

import os
import dataset
import pandas as pd
```

```python
# IMPPORTING TRAINING DATASET

dir = os.path.dirname(os.getcwd())
path = os.path.join(dir,"Datasets","Dataset.csv")
data = pd.read_csv(path)
data
```

```python
# CREATING FEATURE VECTOR
# SECOND PARAAMETER IS 0 FOR LABELLED DATA AND 1 FOR UNLABELLED DATA

f_vec = dataset.f_vector(data.values.tolist(),0)
```

```python
# VISUALISING FEATURE VECTOR

data = pd.DataFrame(f_vec,columns=['dist_d','dist_D','ang_th','H-Bond'])
data
```

```python
# IMPORTING SCI-KIT LIBRARY

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
# DEFINING FEATURES AND LABELS FOR TRAINING THE MODEL

X = data[['dist_d','dist_D','ang_th']]
Y = data[['H-Bond']]
```

```python
# SPLITTING THE EXAMPLES INTO TRAINING AND TESTING DATASETS

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
```

```python
# IMPLEMENTING K-NEAREST NEIGHBORS AND EVALUATING ITS ACCURACY

# Making predictions using scikit learn
from sklearn.neighbors import KNeighborsClassifier

# Create an instance and fit the model 
KNN_Classifier = KNeighborsClassifier(n_neighbors = 5, p = 2, metric= 'minkowski')
KNN_Classifier.fit(x_train, y_train.values.ravel())

# Making predictions
pred_test = KNN_Classifier.predict(x_test)

# Accuracy
print(f"Accuracy = {accuracy_score(y_test, pred_test)}")
```

```python
# IMPLEMENTING LOGISTIC REGRESSION AND EVALUATING ITS ACCURACY

# Making predictions using scikit learn
from sklearn.linear_model import LogisticRegression

# Create an instance and fit the model 
lr_model = LogisticRegression(max_iter = 500)
lr_model.fit(x_train, y_train.values.ravel())

# Making predictions
pred_test = lr_model.predict(x_test)

# Accuracy
print(f"Accuracy = {accuracy_score(y_test, pred_test)}")
```

```python
# IMPLEMENTING SVM AND EVALUATING ITS ACCURACY

# Making predictions using scikit learn
from sklearn.svm import SVC

# Create an instance and fit the model 
svclassifier = SVC(kernel='linear')
svclassifier.fit(x_train, y_train.values.ravel())

# Making predictions
pred_test = svclassifier.predict(x_test)

# Accuracy
print(f"Accuracy = {accuracy_score(y_test, pred_test)}")
```

```python
# EXPORTING THE MODEL

import joblib

joblib.dump(svclassifier,'SVM_TrainedModel.pkl')
```
