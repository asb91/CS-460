```python
# IMPORTING REQUIRED MODULES

import os,sys
import math
import parser
import dataset
import pandas as pd
```

```python
# IMPORTING THE TRAINED MODEL

import joblib

file = open(r'model.txt','r') 
Model = joblib.load(file.read())
file.close()
```

```python
# READING FILES

dir = os.path.dirname(os.getcwd())
path = os.path.join(dir,"Data","Predict")
dirs = os.listdir(path)
```

```python
data = dataset.database(dirs,path)
```

```python
# VISUALISING FEATURE VECTOR

f_vec = data[1]
data_fvec = pd.DataFrame(f_vec,columns=['dist_d','dist_D','ang_th'])
data_fvec
```

```python
# MAKING PREDICTIONS

predictions = Model.predict(data_fvec)
```

```python
# VISUALISING PREDICTIONS

Y = pd.DataFrame(predictions,columns=['Predictions'])
Y
```

```python
# APPENDING PREDICTIONS TO THE DATASET AND VISUALISING IT

data_val = pd.DataFrame(data[0],columns = ['File','S','S_Res','S_RID','S_X', 'S_Y','S_Z','H','H_Res','H_RID','H_X','H_Y','H_Z','N','N_Res','N_RID','N_X','N_Y','N_Z'])
pd.merge(data_val, Y, left_index=True, right_index=True)
```
